Numpy

```python
import numpy as np
heights = np.array([160, 155, 172, 165, 180, 158, 175, 163, 171, 168])
```

Calculate the mean, median, and standard deviation of the students' heights.

```python
heights_mean=np.mean(heights)
print("mean height: ", heights_mean)
heights_median=np.median(heights)
print("median height: ",heights_median)
heights_stdDev=np.std(heights)
print("andatrd deviationb of height: ",heights_stdDev)
                      
```

Create a new array with the heights normalized, i.e., each height value should be transformed to have a mean of 0 and a standard deviation of 1

```python
standardized_heights=(heights-heights_stdDev)/heights_mean
standardized_heights
```

Determine the tallest and shortest students in the class using NumPy's argmax and argmin functions

```python
min_height=np.argmin(heights)
print("index of min height: ", min_height)

max_height=np.argmax(heights)
print("index of max height: ", max_height)
```

Calculate the difference between the tallest and shortest students' heights.

```python
height_differnce=heights[max_height]-heights[min_height]
print("height difference: ", height_differnce)
```

Randomly select 5 heights from the dataset without replacement.

```python
random_heights=np.random.choice(heights,5,replace=False)
print("five random heights from set: ",random_heights)
```

```python
import pandas as pd
# Load dataset
file_path = 'C:\\Users\\swere\\OneDrive - UAP Holdings LTD\\Documents\\KaggleDataSets\\archive\\archive\\Video_Games.csv'
data = pd.read_csv(file_path)
data.head(10)
```

```python
print(data.isna().sum())
data_without_nulls = data.dropna()
print(data_without_nulls.isna().sum())
```

5 Display the top 5 video games by global sale

```python
salesByTopFiveGRegions = data_without_nulls.nlargest(5, 'Global_Sales')
print(salesByTopFiveGRegions)
```

```python
ActionSales = data_without_nulls[data_without_nulls['Genre'] == 'Action']['NA_Sales'].sum()
print("North America total action sales: ",ActionSales)
```

```python
TopPublishers = data_without_nulls.groupby('Publisher')['Global_Sales'].sum().nlargest(3)
print(TopPublishers)
```

```python
shooterSales = data_without_nulls[data_without_nulls['Genre'] == 'Shooter']['EU_Sales'].mean()
print("Shooter sales in Europe: ",shooterSales)
```

```python
JapanSportSales = data_without_nulls[data_without_nulls['Genre'] == 'Sports'].nlargest(1, 'JP_Sales')
print(JapanSportSales)
```

```python
import pandas as pd
# Load dataset
file_path = 'C:\\Users\\swere\\OneDrive - UAP Holdings LTD\\Documents\\KaggleDataSets\\archive\\archive\\jumta.csv'
jumta = pd.read_csv(file_path)
jumta.head(10)
```

```python
# Line plot of monthly revenue
plt.figure(figsize=(10, 5))
plt.plot(jumta['Month'], jumta['Revenue'], marker='o', color='purple', linestyle='-', linewidth=2, label='Revenue')
plt.title('Monthly Revenue (2020/21)')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)
data.tail()
```

```python
fig = px.bar(data, x='Sex', y='Survived', color='Survived', text='Survived', barmode='group', 
             labels={'Survived': 'Number of Passengers'},
             title='Survival Rate by Gender')
fig.show()
```

```python
fig = px.histogram(data, x='Age', color='Survived', nbins=20, 
                   labels={'Age': 'Age of Passengers', 'count': 'Number of Passengers'},
                   title='Age Distribution of Passengers by Survival Status')
fig.show()
```

```python
fig = px.violin(data, x='Pclass', y='Fare', box=True, points='all', 
                labels={'Pclass': 'Passenger Class', 'Fare': 'Fare Price'},
                title='Fare Price Distribution by Passenger Class')
fig.show()
```

###### fig = px.scatter(data, x='Age', y='Fare', color='Pclass', symbol='Pclass',
                 labels={'Age': 'Age of Passengers', 'Fare': 'Fare Price', 'Pclass': 'Passenger Class'},
                 title='Age vs. Fare Price by Passenger Class')
fig.show()

```python
fig = px.scatter_matrix(data)
```

```python
# Export your final visualizations as HTML files to share with others.
import plotly.io as pio

pio.write_html(fig, 'bar_plot.html')
pio.write_html(fig, 'histogram_plot.html')
pio.write_html(fig, 'violin_plot.html')
pio.write_html(fig, 'scatter_plot.html')
pio.write_html(fig, 'scatter_matrix.html')
```
